```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
#%matplotlib inline
```

#LAW OF LARGE NUMBERS
The law of large numbers states that as a sample size grows, the sample mean gets closer to the population mean.

```
number_of_experiments = 10000
population_mean = 3.5 #expectation of a fair die
error = []
for exp in range(1, number_of_experiments+1):
    dice_rolles = np.random.randint(1, 7, exp)
    sample_mean = dice_rolles.mean()
    error.append(abs(sample_mean - population_mean))
fig = plt.figure()    
plt.plot(np.arange(1, number_of_experiments), np.array(error)[1:])
plt.xlabel('Sample Size')
plt.ylabel('Error')
#plt.show()
```

#CENTRAL LIMIT THEOREM
The central limit theorem states that if you have a population with mean μ and standard deviation σ and take sufficiently large random samples from the population with replacement, then the distribution of the sample means will be approximately normally distributed. This will hold true regardless of whether the source population is normal or skewed, provided the sample size is sufficiently large (usually n > 30). http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/BS704_Probability/BS704_Probability12.html
